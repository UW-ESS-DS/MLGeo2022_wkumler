```python
import numpy as np
import matplotlib.pyplot as plt

this_list = [1, 2, 3, 5, 8]

now_array = np.array(this_list)

new_array = np.arange(10)

gauss_arr = np.random.randn(1000)

plt.hist(gauss_arr)
```

```python
def fancyhist(num_arr):
    fig, ax = plt.subplots(figsize=(6.5, 4))
    plt.hist(num_arr, bins=np.arange(-4, 4, 0.25), 
             color="lightseagreen")
    
    ax.grid(color="white", )
    # ax.grid(color="grey")
    # ax.set_axisbelow(True)
    
    ax.set_xlabel("Normally distributed values", fontsize=16)
    ax.set_ylabel("Counts", fontsize=16)
    plt.title('Customized histogram')
    ax.title.set_fontsize(20)
    plt.show()
    plt.savefig('fancyhist.png',dpi=300)

fancyhist(gauss_arr)
```
